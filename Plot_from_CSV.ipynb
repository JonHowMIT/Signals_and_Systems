```python
data = []
import csv
import numpy as np
with open('data/file_2023-07-10_24.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        frow = [float(jj) for jj in row]
        if (len(data) == 0):
            data = frow
        else:
            data = np.r_['0,2,1', data, frow]
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib import rcParams
rcParams["font.serif"] = "cmr14"
rcParams['savefig.dpi'] = 300
rcParams["figure.dpi"] = 100
rcParams.update({'font.size': 18})
rcParams['axes.grid'] = True
rcParams['lines.linewidth'] = 2.0
params = {'legend.fontsize': 12,
          'legend.handlelength': 2}
plt.rcParams["figure.figsize"] = (8,5)
plt.rcParams.update(params)
```

```python
fig, ax = plt.subplots(figsize=(16,8))
Npts = 6000
plt.plot(data[-Npts:,0],data[-Npts:,1]*180/np.pi,'b-',label='theta [rad]')
#plt.plot(data[-Npts:,0],data[-Npts:,2],'r--',label='gyro [rad/s]')
plt.plot(data[-Npts:,0],data[-Npts:,3],'g:',label='U/100')
plt.legend(loc=3)
plt.show()
```
