# Simple demonstration of what Amplitude Modulation Actually looks like

```python
from os.path import exists
file_exists = exists('basic_material.py')

if (not file_exists):
    !wget -O basic_material.py https://raw.githubusercontent.com/JonHowMIT/Signals_and_Systems/main/basic_material.py
%run basic_material

from scipy import signal
```

```python
t_dense = np.linspace(0,2,10**5)  #       
```

```python
sig = np.sin(2 * np.pi * t_dense)
y_signal = 0.5 + 0.25*signal.square(2 * np.pi * 10 * t_dense, duty=(sig + 1)/2)
y_carrier = np.cos(2*np.pi*5000*t_dense)
y_modulated = y_signal*y_carrier
```

```python
fig, axs = plt.subplots(3,1,figsize=(5, 3),dpi=150,constrained_layout = True)
axs[0].plot(t_dense,y_signal,'b-',lw=2,label='sig')
axs[1].plot(t_dense,y_carrier,'r-',lw=2,label='car')
axs[2].plot(t_dense,y_modulated,'g-',lw=2,label='mod')
for ii in np.arange(3):
    axs[ii].legend()
    axs[ii].set_xlim([0,1.5])
    nicegrid(axs[ii])
nicegrid()
axs[-1].set_xlabel("Time [s]")
plt.savefig('./figs/AM_1.png', bbox_inches='tight', dpi=300)
plt.show()
```
