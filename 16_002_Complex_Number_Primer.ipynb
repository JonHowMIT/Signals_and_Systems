## <a name="s1" style="color:blue">Various Numbers</a>
### Complex Numbers

We will denote the square root of $-1$ by $j$ (though frequently you will see $i$ used as well), that is, $j^2 = -1$. 

An *imaginary number* is one of the form $\beta j$, where $\beta$ is a real number. 

A *complex number* is one of the form $\alpha + \beta j$, where $\alpha$ and $\beta$ are real numbers.

### Real and imaginary parts

For a complex number $s = \alpha + \beta j$, where $\alpha$ and $\beta$ are real numbers, the real and imaginary parts are given by 
\begin{equation}
\mathrm{Re}[s] = \alpha; ~~ \mathrm{Im}[s] = \beta.
\end{equation}
The real and imaginary parts of a complex number $s$ are sometimes denoted $\Re[s]$ and $\Im[s]$, respectively. Two complex numbers are equal if, and only if, the real parts are equal to each other and the imaginary parts are also equal.

## Absolute value of a complex number
Similar to a 2D vector, the \emph{absolute value of a complex number} is given by
\begin{equation}
|s| = r = |\alpha + \beta j| = \sqrt{\alpha^2 + \beta^2}.
\end{equation}
The absolute value of a complex number (also known as its magnitude) is a non-negative number, and is zero only if the number is zero (that is, the real and imaginary parts are both zero). 

## Phase of a complex number
The angle between the horizontal axis and the complex number (positive in a counter clockwise direction, by convention), as denoted on the complex plane, is known as the *phase* of the complex number. <br> 
It is sometimes also called the *argument* or *angle* of the complex number (denoted $\theta$ in Figure 1). The counterclockwise direction denotes the positive angle. 

The phase of a complex number $s = \alpha + \beta j$, denoted $\angle{s}$, is given by
\begin{equation}
\angle{s} = \angle{(\alpha + \beta j)} = \arctan \frac{\beta}{\alpha}
\end{equation}
Remember that angles that differ by $2\pi k$, where $k$ is an integer, are considered equal. <br>
If the angle is zero ($\angle{s}=0$), then $s$ is a real number.

## <a name="s3" style="color:blue">Complex exponential representation</a>

As seen in Figure 2, a complex number $s = \alpha + \beta j$ can be alternatively expressed as
\begin{equation}
s = r e^{j\theta},
\end{equation}
where $r=|s|$ and $\theta=\angle s$. This is called the \emph{polar coordinate} representation.

### Conjugate
The conjugate of a complex number  $s = \alpha + \beta j$, denoted $s^*$, is 
\begin{equation}
s^* = \alpha - \beta j. 
\end{equation}
In the complex plane (see Figure~2), $s^*$ is obtained by reflecting $s$ through the horizontal axis. <br>
In the complex exponential representation, we note that
\begin{equation}
s^* = r e^{-j\theta}.
\end{equation}

### Euler's formula
For any real number, $\theta$, Euler's formula states that 
\begin{equation}
e^{j\theta} = \cos \theta + j \sin \theta,
\end{equation}
which can also be seen from Figure \ref{fig:cxplane} for the case of a complex number $s$ that lies on the unit circle, that is, $|s| = r =1$. 

## <a name="s4" style="color:blue">Operations</a>

### Addition and subtraction of Complex numbers
The sum of two complex numbers is obtained by separately adding the real and imaginary parts. Subtraction is similarly defined.  In other words,
\begin{eqnarray}
s_1 + s_2 &=& (\alpha + \beta j) + (\gamma + \delta j) = (\alpha + \gamma) + (\beta + \delta) j \\
s_1 - s_2 &=& (\alpha + \beta j) - (\gamma + \delta j) = (\alpha - \gamma) + (\beta - \delta) j
\end{eqnarray}

The real and imaginary parts of a complex number can be expressed in terms of the number and its conjugate as follows:
\begin{equation}
\mathrm{Re}[s] = \alpha = \frac{s + s^*}{2}; ~~ \mathrm{Im}[s] = \beta = \frac{s - s^*}{2j}.
\end{equation}

### Multiplication of Complex numbers
The product of two complex numbers is given by 
\begin{eqnarray}
s_1 s_2 &=& (\alpha + \beta j)(\gamma + \delta j) = (\alpha \gamma - \beta \delta) + (\beta\gamma + \alpha\delta) j. 
\end{eqnarray}
From the above, we note that the product of a complex number and its conjugate is the square of its magnitude, that is:
\begin{eqnarray}
s s^* &=& \alpha^2 + \beta^2  = |s|^2.
\end{eqnarray}

In the exponential form, the product of two complex numbers can be written as
\begin{eqnarray}
s_1 s_2 &=& (r_1e^{j\theta_1}) (r_2e^{j\theta_2}) = r_1r_2 e^{j(\theta_1+ \theta_2)}. 
\end{eqnarray}
In other words (see Figure~3), the magnitudes multiply and the phases add. <br>

<b> Examples: </b> 
The above fact can also be used to show that the product of a complex number and its conjugate is the square of its magnitude. 
$$
s s^* = |s|^2
$$
In addition, note that multiplication by a real number scales the number, 
$$
a s = a (r e^{j\theta_1}) = (a r) e^{j\theta_1}
$$
and multiplication by $e^{j\theta}$ rotates the number by $\theta$ on the complex plane. 
$$
s e^{j\theta} = r e^{j\theta_1} e^{j\theta} = r e^{j(\theta_1 + \theta)}
$$
