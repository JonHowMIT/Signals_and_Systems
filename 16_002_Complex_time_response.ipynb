```python
a = 4
b = -2
c = -3
d = 8
AB = complex(a,b)
CD = complex(c,d)

phi = np.arctan2(b,a)
r = np.sqrt(a**2+b**2)
```

```python
t = np.linspace(0,2,1000)

x = AB*np.exp(CD*t) + np.conjugate(AB)*np.exp(np.conjugate(CD)*t)
x_simplified = 2*np.exp(c*t)*(a*np.cos(d*t) - b*np.sin(d*t))
x_condensed = 2*r*np.exp(c*t)*np.cos(d*t+phi)
```

```python
fig = figure(figsize=(8, 5),dpi=150,constrained_layout = True)
plt.plot(t,x.real,'b-',lw=5,label='Real(x(t))')
plt.plot(t,x.imag,'y-',lw=3,label='Imag(x(t))')
plt.plot(t,x_simplified,'r-',label='x(t) simplified')
plt.plot(t,x_condensed,'g:',lw=4,label='x(t) condensed')
plt.legend(loc='best')
plt.show()
```
